```python
import tensorflow as tf
import numpy as np
```

```python
DATA_DIR='shakespeare.txt'
BATCH_SIZE=25
LAYER_NUM=2
SEQ_LENGTH=75
HIDDEN_DIM=128
GENERATE_LEN=500
NB_EPOCH=1500
MODE='train'
WEIGHTS=''
LEARNING_RATE=2e-3
IS_TRAIN=True
```

```python
def get_text_data():
    with open(DATA_DIR,"r") as text_file:
        data=text_file.read()
        tr_data,va_data=data[0:1070392],data[1070392:1075392]
    lis=list(set(data))    
    total_char,unique_char=len(data),len(lis)
    print ('data has %d characters, %d unique.' %(total_char, unique_char))
    char_to_ix = { ch:i for i,ch in enumerate(lis) }
    ix_to_char = { i:ch for i,ch in enumerate(lis) }
    return data

def reset_graph(seed=42):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)
```

```python
data=get_text_data()
```

```python
import json

with open('coding2.json') as file:
    da=json.load(file)
    total_chars,unique_chars,char_id,id_char=da['total'],da['unique'],da['char_id'],da['id_char']

id_v={int(k):v for k,v in id_char.items()}
```

```python
id_v
```

```python
def generate_next_batch(given_data, seq_length, p):
    inputs = [char_id[ch] for ch in given_data[p:p+seq_length]]
    targets = [char_id[ch] for ch in given_data[p+1:p+seq_length+1]]
    inputs=np.array(inputs).reshape(1,seq_length)
    targets=np.array(targets).reshape(1,seq_length)
    return inputs,targets
```

```python
ptr=0
batches=(total_chars//SEQ_LENGTH)
x=[]
y=[]
for batch in range(batches):
    a,b=generate_next_batch(data, SEQ_LENGTH, ptr)
    x.append(a)
    y.append(b)
    ptr+=SEQ_LENGTH
x=np.array(x).reshape(batches, SEQ_LENGTH)
y=np.array(y).reshape(batches, SEQ_LENGTH)
```

```python
#test generator function here
def generator(seed_i,n):
    x= seed_i
    ixes=[]
    ixes.append(seed_i[0][0])
    global IS_TRAIN
    IS_TRAIN=False
    for t in range(n):
        p=sess.run(prediction,{X:x})
        ix = np.random.choice(range(unique_chars), p=p[0].ravel())
        x[0][0]=ix
        ixes.append(ix)
    return ixes
```

```python
def build_LSTM(x):
    with tf.name_scope('lstm'):
        cells= [tf.contrib.rnn.BasicLSTMCell(num_units=HIDDEN_DIM )
            for layer in range(LAYER_NUM)]
        multi_layer_cell = tf.contrib.rnn.MultiRNNCell(cells, state_is_tuple=True)
        dropout=tf.nn.dropout(x,0.80)
#         batch_size_T  = np.shape(x)[0]
        _initial_state =  multi_layer_cell.zero_state(tf.shape(x)[0], tf.float32)
        
        outputs, states = tf.nn.dynamic_rnn(multi_layer_cell, x,initial_state=_initial_state, dtype=tf.float32)
        return outputs,states
```

```python
reset_graph()

# global SEQ_LENGTH
# if(IS_TRAIN):
#   SEQ_LENGTH=25
# else:
#   SEQ_LENGTH=1
X=tf.placeholder(tf.int32,[None,SEQ_LENGTH])
Y=tf.placeholder(tf.int32,[None,SEQ_LENGTH])

X_onehot=tf.one_hot(X,unique_chars)
Y_onehot=tf.one_hot(Y,unique_chars)

X_seqs = tf.unstack(X,axis=1)

outputs,states=build_LSTM(X_onehot)

W=tf.Variable(tf.truncated_normal((HIDDEN_DIM,unique_chars),-1,1))
B=tf.Variable(tf.ones((1,unique_chars)))

flat_outputs = tf.reshape(tf.concat(axis=1, values=outputs), [-1, HIDDEN_DIM])
flat_y = tf.reshape(tf.concat(axis=1, values=Y_onehot), [-1])
Ys=tf.matmul(flat_outputs,W)+B

prediction=tf.nn.softmax(Ys)
xentropy = tf.nn.softmax_cross_entropy_with_logits_v2(labels=flat_y, logits=Ys)
loss = tf.reduce_mean(xentropy)

with tf.name_scope('adam_optimizer'):
    optimizer = tf.train.AdamOptimizer(LEARNING_RATE)
    grad_vars=optimizer.compute_gradients(loss)
    capped_gvs = [(tf.clip_by_value(grad, -5., 5.), var) for grad, var in grad_vars]
    minimize = optimizer.apply_gradients(capped_gvs)
```

```python
saver=tf.train.Saver()
init=tf.global_variables_initializer()

no_of_batches=int(total_chars//BATCH_SIZE)
ptr=0
va_ptr=0
```

```python
with tf.Session() as sess:
    smooth_loss_tr = -np.log(1.0/unique_chars)*BATCH_SIZE # loss at iteration 0
    smooth_loss_va = -np.log(1.0/unique_chars)*BATCH_SIZE # loss at iteration 0
    sess.run(init)
    for i in range(NB_EPOCH):
        for j in range(batches//BATCH_SIZE):
            if (ptr+BATCH_SIZE+1 >= batches): 
                ptr=0
#             if va_ptr+BATCH_SIZE+1>=len(va_data):
#               va_ptr=0
            inputs,targets=x[ptr:ptr+BATCH_SIZE],y[ptr:ptr+BATCH_SIZE]
            ptr+=BATCH_SIZE
            if(j%100==0):
#               va_in,va_ta=generate_next_batch(va_data, SEQ_LENGTH, va_ptr)
#               va_ptr=va_ptr+BATCH_SIZE
              global IS_TRAIN
              IS_TRAIN=True
              loss_t=sess.run(loss,feed_dict={X:inputs, Y:targets}) 
#               loss_v=sess.run(loss, feed_dict={X:va_in, Y:va_ta})
              sample_ix=generator(np.array(inputs),200)
              txt = ''.join(id_v[ix] for ix in sample_ix)
              smooth_loss_tr = smooth_loss_tr * 0.999 + loss_t * 0.001
#               smooth_loss_va = smooth_loss_va * 0.999 + loss_v * 0.001
              print('epoch %i iteration %i loss %f va'%(i,j,smooth_loss_tr))
              print ('----\n %s \n----' % (txt, ))
            sess.run(minimize,feed_dict={X:inputs,Y:targets})
        if(i%10==0):
            save_path = saver.save(sess, "lr-new/LSTMmodel-new"+str(i)+"loss"+str(int(smooth_loss_tr))+".ckpt")
```
